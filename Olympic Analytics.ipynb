```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("athlete_events.csv")
```

```python
df1=pd.read_csv("noc_regions.csv")
```

```python
atheletes_df = df.merge(df1,how='left',on="NOC")
atheletes_df.head()
```

```python
top10 = atheletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top10
```

```python
sns.barplot(x=top10.index,y=top10)
plt.title("Top 10 Participating Countries")
plt.xticks(rotation=20)
plt.xlabel("Countries")
plt.show()
```

```python
plt.hist(atheletes_df.Age,bins=np.arange(10,80,2),color="Orange",edgecolor="white")
plt.title("Age Distribution of the Participants")
plt.xlabel("Age (yrs)")
plt.ylabel("Number of Participants")
plt.show()
```

```python
winter_sport = atheletes_df[atheletes_df.Season == 'Winter'].Sport.unique()
winter_sport
```

```python
summer_sport = atheletes_df[atheletes_df.Season == 'Summer'].Sport.unique()
summer_sport
```

```python
gender=atheletes_df.Sex.value_counts()
gender
```

```python
 plt.pie(gender,labels=gender.index,autopct='%1.1f%%')
plt.title("Gender Distribution")
plt.show()
```

```python
medal=atheletes_df.Medal.value_counts()
medal
```

```python
female_athelete = atheletes_df[(atheletes_df.Sex == "F") & (atheletes_df.Season == "Summer")][['Sex','Year']]
female_athelete = female_athelete.groupby('Year').count().reset_index()
female_athelete.tail()
```

```python
women = atheletes_df[(atheletes_df.Sex == "F") & (atheletes_df.Season == "Summer")]
plt.figure(figsize=(20,10))
plt.title("Women Participation")
sns.countplot(x="Year",data=women)
```

```python
max_year=atheletes_df.Year.max()
max_year
```

```python
team_names = atheletes_df[(atheletes_df.Year == max_year) & (atheletes_df.Medal == "Gold")].Team
team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(10),y=team_names.value_counts().head(10).index)
plt.ylabel(None)
plt.xlabel("Country-wise Medal for the Year 2016")
plt.show()
```

```python
not_null_medals = atheletes_df[(atheletes_df['Height'].notnull()) & (atheletes_df['Weight'].notnull())]
```

```python
sns.scatterplot(x="Height",y="Weight",data=not_null_medals,hue="Sex")
plt.title("Height and Weight of Olympic Medalist")
plt.show()
```
